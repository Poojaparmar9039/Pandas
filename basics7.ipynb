```python
customer=pd.DataFrame({
        'customerid': [1, 2, 3],
        'name': ['riya', 'tiya', 'siya']
    })
```

```python
orders=pd.DataFrame({
    'customerid':[1,2,4],
    'amount':[100,200,300]
})
```

```python
merged=pd.merge(customer,orders,on='customerid',how="inner")
print(merged)
```

```python
merged=pd.merge(customer,orders,on='customerid',how="left")
print(merged)
```

```python
merged=pd.merge(customer,orders,on='customerid',how="outer")
print(merged)
```

```python
merged=pd.merge(customer,orders,on='customerid',how="right")
print(merged)
```

```python
merged=pd.merge(customer,orders,on='customerid',how="cross")
print(merged)
```
