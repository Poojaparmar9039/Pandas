NaN (Not a number)

None(for object data types)

isnull()
True---Nan is missing
False--value is present


```python
import pandas as pd
```

```python
data = {
    "Name": [
        "Amit", None, "Rahul", "Pooja", None,
        "Anjali", "Rohit", "Sneha", "Karan", "Priya"
    ],
    "Age": [
        25,  None, 30, 24,  None,
        27, 32, 26, 29, 31
    ],
    "Salary": [
        35000, None, 55000, 30000,  None,
        48000, 60000, 40000, 52000, 58000
    ],
    "Performance_Score": [
        78, None, 90, 72,  None,
        82, 91, 80, 86, 89
    ]
}

df=pd.DataFrame(data)
print(df)
```

```python
print(df.isnull())
```

```python
#dropna----clear none values
#axis=0  --drop rows of missing value while 1 will drop colums
df.dropna(axis=0,inplace=True)
print(df)
```

```python
data = {
    "Name": [
        "Amit", None, "Rahul", "Pooja", None,
        "Anjali", "Rohit", "Sneha", "Karan", "Priya"
    ],
    "Age": [
        25,  None, 30, 24,  None,
        27, 32, 26, 29, 31
    ],
    "Salary": [
        35000, None, 55000, 30000,  None,
        48000, 60000, 40000, 52000, 58000
    ],
    "Performance_Score": [
        78, None, 90, 72,  None,
        82, 91, 80, 86, 89
    ]
}

df=pd.DataFrame(data)
print(df)
```

```python
#fillna(default_value,inplace=True)
#fill the none value

df.fillna(0,inplace=True)
print(df)

```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())

print(df)
```
