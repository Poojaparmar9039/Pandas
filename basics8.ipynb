## concatenation

vertically(row)  or horizontally(column)

```python
import pandas as pd
```

```python
customer=pd.DataFrame({
        'customerid': [1, 2, 3],
        'name': ['riya', 'tiya', 'siya']
    })
```

```python
orders=pd.DataFrame({
    'customerid':[5,6,4],
    'name':['ali','shashank','deepa']
})
```

```python
concat_df=pd.concat([customer,orders],ignore_index=True)
print(concat_df)
```

```python
concat_df=pd.concat([customer,orders],axis=1,ignore_index=True)
print(concat_df)
```
