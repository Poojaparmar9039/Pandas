```python
data = {
    "Name": [
        "Amit", "Neha", "Rahul", "Pooja", "Vikas",
        "Anjali", "Rohit", "Sneha", "Karan", "Priya"
    ],
    "Age": [
        25, 28, 30, 24, 35,
        27, 32, 26, 29, 31
    ],
    "Salary": [
        35000, 42000, 55000, 30000, 65000,
        48000, 60000, 40000, 52000, 58000
    ],
    "Performance_Score": [
        78, 85, 90, 72, 88,
        82, 91, 80, 86, 89
    ]
}

df=pd.DataFrame(data)
print("sample data ")
print(df)
print("discriptive statistics")
print(df.describe())
```

```python
high_Salary=df[(df["Salary"]>50000) & (df["Age"]>30)]
print(high_Salary)
high_Salary=df[(df["Salary"]>50000) | (df["Age"]>30)]
print(high_Salary)
```
