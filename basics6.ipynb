```python
import pandas as pd
```

```python
data={
    "Name":['Arun','Varun','karan','jack','john'],
    "Age":[14,10,30,10,35],
    "Salary":[100,200,300,200,6000]
}

df=pd.DataFrame(data)
```

```python
avg_salary=df['Salary'].mean()
```

```python
avg_salary
```

Grouping

```python
grouped=df.groupby('Age')["Salary"].sum()
print(grouped)
```

```python
grouped=df.groupby('Age')["Salary"].min()
print(grouped)
```
